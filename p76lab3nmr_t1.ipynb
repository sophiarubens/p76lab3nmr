```python
import numpy as np
import math
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit

def t1_min(t,T1):
    return M0*(1-2*np.exp(-t/T1))

T1bound=30 
# use the same bound for all
# expect all T1 values to be < 30 s for these materials
# confirmed no effect on result for 30, 15, 10, 5, 1, 0.1, 0.05 for mineral oil
```

```python
#HEAVY MINERAL OIL
min_t=np.array([ 0.01,  0.015, 0.020, 0.025, 0.030, 0.035, 0.040, 0.045, 0.05 ])
min_m=np.array([ -57.6, -38.4, -22.4, -2.40, 2.40,  16.8,  27.2,  34.4,  41.6 ])
time_err=0.00001
mag_err=0.1
M0=106

min_T1,min_T1_var=curve_fit(t1_min,min_t,min_m,bounds=(0,T1bound))
plt.scatter(min_t,min_m,label='data',color='b')
plt.errorbar(min_t,min_m,mag_err,time_err,linestyle='',marker='s',mfc='r',mec='r',ms='3')
plt.plot(min_t,t1_min(min_t,*min_T1),label='fit')
plt.xlabel('tau (s)')
plt.ylabel('FID amplitude (mV)')
plt.title('mineral oil T_1 fit')
plt.text(0.025,-40,'M(tau)=106*(1-2*exp(-tau/0.04034978))')
plt.legend()
plt.show()
```
